```python
from apify_client import ApifyClient

client = ApifyClient("apify_api_biqQYOhWEwWaZFC8HjsAiwt5St34tH1bpruh")

# Prepare the Actor input
run_input = {
    "startUrls": [{ "url": "https://www.tripadvisor.fr/Hotel_Review-g187147-d16711418-Reviews-Ibis_Styles_Paris_Meteor_Avenue_D_Italie-Paris_Ile_de_France.html" }],
    "maxItemsPerQuery": 20,
    "reviewRatings": ["ALL_REVIEW_RATINGS"],
    "reviewsLanguages": ["ALL_REVIEW_LANGUAGES"],
}

# Run the Actor and wait for it to finish
run = client.actor("maxcopell/tripadvisor-reviews").call(run_input=run_input)

# Fetch and print Actor results from the run's dataset (if there are any)
print("üíæ Check your data here: https://console.apify.com/storage/datasets/" + run["defaultDatasetId"])
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    print(item)
```
